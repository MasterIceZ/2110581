```python
def check_subset(l1, l2) :
  ok = True
  j = 0
  for i in l1 :
    while (l2[j] < i) : 
      j += 1
    if (l2[j] != i) :
      ok = False
      break
    j += 1
  return ok
```

```python
def calculate_mass(peptide):
    return sum(mass[aa] for aa in peptide)

def score(peptide, spectrum):
  peptide_mass = [0]
  for aa in peptide:
      peptide_mass.append(peptide_mass[-1] + mass[aa])
  peptide_spectrum = [0] + sorted(peptide_mass[1:] + [peptide_mass[-1] - m for m in peptide_mass[:-1]])
  return sum(1 for mass in peptide_spectrum if mass in spectrum)

def trim_leaderboard(leaderboard, spectrum, N):
    scored_peptides = [(peptide, score(peptide, spectrum)) for peptide in leaderboard]
    scored_peptides.sort(key=lambda x: x[1], reverse=True)
    return [peptide for peptide, _ in scored_peptides[:N]]

def leaderboard_cyclopeptide_sequencing(spectrum, N):
  leaderboard = [""]
  leader_peptide = ""
  while leaderboard:
      leaderboard = [peptide + aa for peptide in leaderboard for aa in mass]
      leaderboard = [peptide for peptide in leaderboard if calculate_mass(peptide) <= max(spectrum)]
      for peptide in leaderboard:
          if calculate_mass(peptide) == max(spectrum):
              if score(peptide, spectrum) > score(leader_peptide, spectrum):
                  leader_peptide = peptide
      leaderboard = trim_leaderboard(leaderboard, spectrum, N)
  return leader_peptide

best_peptide = leaderboard_cyclopeptide_sequencing(spectrum, 13)
print("Best Peptide:", best_peptide)
print(check(spectrum, best_peptide))
```
